This notebook provides functionality to convert any Hugging Face model to GGUF format and optionally upload it back to Hugging Face.

```python
from huggingface_hub import snapshot_download

model_id="thuanan/Llama-3.2-1B-Instruct-Chat-sft"
snapshot_download(repo_id=model_id, local_dir="llama-hf-sft", revision="main")
```

```python
!python llama.cpp/convert_hf_to_gguf.py llama-hf-sft \
  --outfile Llama-3.2-1B-Instruct-Chat-sft.gguf \
  --outtype bf16
```

```python
from huggingface_hub import HfApi
api = HfApi()

model_id = "thuanan/Llama-3.2-1B-Instruct-Chat-sft-gguf"
api.create_repo(model_id, exist_ok=True, repo_type="model")
api.upload_file(
    path_or_fileobj="Llama-3.2-1B-Instruct-Chat-sft.gguf",
    path_in_repo="Llama-3.2-1B-Instruct-Chat-sft.gguf",
    repo_id=model_id,
)
```
